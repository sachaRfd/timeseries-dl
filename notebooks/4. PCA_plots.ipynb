# Extra: Plotting the samples in a latent Space: 


```python
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA




train_dataset_path = "../dataset/classification/FingerMovements/train.pt"
train_data = torch.load(train_dataset_path)
samples, labels = train_data["samples"], train_data["labels"]

# mean the samples to plot them on PCA: 
samples_meaned = samples.mean(dim=1)
```

#### Plotting with PCA: 

```python
pca = PCA(n_components=2)

samples_pcaed = pca.fit_transform(samples_meaned.numpy())


# Plot PCA results
plt.figure(figsize=(8, 6))
scatter = plt.scatter(samples_pcaed[:, 0], samples_pcaed[:, 1], c=labels.numpy(), cmap="viridis", alpha=0.7)
plt.colorbar(scatter, label="Labels")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA of Samples")
plt.show()
```
